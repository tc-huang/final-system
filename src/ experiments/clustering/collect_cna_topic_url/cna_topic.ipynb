```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.keys import Keys as KEY
import time
```

```python
driver = webdriver.Chrome()
time.sleep(3)
```

```python
import pandas as pd
df = pd.DataFrame(columns=['title', 'url', 'topic_num'])
```

```python
driver = webdriver.Chrome()
time.sleep(3)
topic_num = 3895

while topic_num > 3600:
    url = f"https://www.cna.com.tw/topic/newstopic/{topic_num}.aspx"
    
    driver.get(url)
    try:
        title = driver.find_element(By.CLASS_NAME, "Temasname").text
        print(topic_num, title)
        # df = df.append({'title': title, 'url': url, 'topic_num': topic_num}, ignore_index=True)
        df = pd.concat([df, pd.DataFrame({'title': [title], 'url': [url], 'topic_num': [topic_num]})], ignore_index=True)
    except:
        pass
    topic_num -= 1
    time.sleep(5)
```
