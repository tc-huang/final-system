```python
# from_dir = "./new_topic_json"
# target_file = "topic4091.json"
# to_dir = "./json-data/"

target_file = "./news-data-2023-02_sample20_perday_01.csv"

import pandas as pd 
df = pd.read_csv(target_file, encoding="utf-8")
```

```python
df
```

```python
import re
import json

def clean_text(text):
    text = re.sub("〔.*?〕","",text)
    return text

def content_to_list(content):
    content = content.replace("{","")
    content = content.replace("}","")
    result = []
    for i in content.split(","):
        if len(i.replace("\n","").replace("\"","").replace("\'","")) != 0:
            result.append(i)
    return result
    
```

```python
result = []

for index, news_data in df[df['chose_to_label'] == True].iterrows():
    news = {
    "data": {
        "title": None,
        "paragraphs": None,
        "uid": None,
        "url": None
    }
}
    news['data']['title'] = news_data['title']
    news['data']['paragraphs'] = []
    content = news_data['content']
    content = clean_text(content)
    content = content_to_list(content)
    news['data']['uid'] = news_data['uid']
    news['data']['url'] = news_data['url']

    for i, paragraph_data in enumerate(content):
        paragraph = {
    "text": None,
    "author": None,
}
        paragraph['text'] = paragraph_data
        paragraph['author'] = i
        news['data']['paragraphs'].append(paragraph)
    result.append(news)
```

```python
result
```

```python
file_name = target_file.replace(".csv",".json")
with open(file_name, "w") as f:
    json.dump(result, f, indent=4, ensure_ascii=False)
```
