```python
df_02 = pd.read_csv('./news-data-2023-02.csv')
```

```python
df_02['source'].value_counts()
```

```python
df_01['time'] = pd.to_datetime(df_01['time'])
df_02['time'] = pd.to_datetime(df_02['time'])

df_01['time']
```

```python
df_01['day'] = df_01['time'].dt.day
df_02['day'] = df_02['time'].dt.day
```

```python
df_01['chose_to_label'] = False
df_02['chose_to_label'] = False
```

```python
sample_size_per_day = 20
random_seed = 777
```

```python
for day in df_01['day'].unique():
    sample_in_data = df_01[(df_01['day'] == day) & (df_01['content'].str.len() >= 10)].sample(n = sample_size_per_day, replace=False, random_state=random_seed)
    sample_in_data['chose_to_label'] = True
    df_01.update(sample_in_data)
```

```python
for day in df_02['day'].unique():
    sample_in_data = df_02[(df_02['day'] == day) & (df_02['content'].str.len() >= 10)].sample(n = sample_size_per_day, replace=False, random_state=random_seed)
    sample_in_data['chose_to_label'] = True
    df_02.update(sample_in_data)
```
